## Load samples file

```python
samples = pd.read_csv (samples_file, comment="#", skip_blank_lines=True, index_col=0)
```

```python
for sample in samples.index[[1]]:
    # Get variables
    input_video = os.path.join(data_dir, sample + ".mp4")
    int_floor = samples.loc[sample, "int_floor"]
    int_ceiling = samples.loc[sample, "int_ceiling"]
    area_floor = samples.loc[sample, "area_floor"]
    area_ceiling = samples.loc[sample, "area_ceiling"]
    vid_length = samples.loc[sample, "vid_length"]
    # Run idtrackerai
    !idtrackerai terminal_mode \
          --_video $input_video \
          --_bgsub 'True' \
          --_intensity [$int_floor,$int_ceiling] \
          --_area [$area_floor,$area_ceiling] \
          --_range [0,$vid_length] \
          --_nblobs 2 \
          --_session $sample \
          --exec track_video
```
