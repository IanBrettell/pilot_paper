```
import json
import numpy as np

false = False
true = True

class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return super(NpEncoder, self).default(obj)

for sample in meta_df.index[0:5]:
  input_video = os.path.join('videos', sample + '.mp4')
  intensity = [meta_df.loc[sample, "int_floor"], meta_df.loc[sample, "int_ceiling"]]
  area = [meta_df.loc[sample, "area_floor"], meta_df.loc[sample, "area_ceiling"]]
  range = [0, meta_df.loc[sample, "vid_length"]]
  data = {
    "open-multiple-files": false,
    "_session": {"value": sample},
    "_video": {"value": input_video},
    "_range": {"value": range, "max": 18566, "min": 0, "scale": 1.0, "convert-int": true},
    "_rangelst": {"value": ""},
    "_multiple_range": {"value": "False"},
    "_intensity": {"value": intensity,"max": 255, "min": 0, "scale": 1.0, "convert-int": true},
    "_area": {"value": area, "max": 60000, "min": 0, "scale": 1.0, "convert-int": true},
    "_nblobs": {"value": 2.0},
    "_resreduct": {"value": 1.0},
    "_chcksegm": {"value": "False"},
    "_roi": {},
    "_no_ids": {"value": "False"},
    "_bgsub": {"value": "True"}
  }

  out_path = os.path.join('videos', sample + '.json')

  with open(out_path, 'w') as outfile:
    json.dump(data, outfile, cls = NpEncoder)
```
